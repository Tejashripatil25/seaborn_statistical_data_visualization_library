```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Datasets

```python
#load datset
i= sns.load_dataset('iris')
i.head()
```

```python
#load titanic dataset
t = sns.load_dataset('titanic')
t.head()
```

### Plotting using Seaborn
#### Histogram using distance plotting

#### Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```python
#Flexibly plot a univariate distribution of observations.
sns.histplot(data=t,)
```

```python
sns.histplot(data=t[['age','fare']])
```

```python
#histplot
#lets supply iris flower dataset
sns.histplot(data=i)
```

```python
#lets plot histogram for one column
sns.histplot(data=i['sepal_width'])
```

```python
type(i)
```

```python
#lets supply own range
sns.histplot(data=i['sepal_width'],bins=[1.5,2.5,3.6,4.3,5])
```

```python
#lets set new labels
sns.histplot(data=i['sepal_width'],bins=[1.5,2.5,3.6,4.3,5])
plt.xlabel('Sepal Lenght in cm')
plt.ylabel('Rangewise count')
plt.title('Sample plot')
```

```python
#lets set new labels: using seaborn
ax = sns.histplot(data=i['sepal_width'],bins=[1.5,2.5,3.6,4.3,5])
ax.set_xlabel('Sepal Lenght in cm')
ax.set_ylabel('Rangewise count')
ax.set_title('Sample Plot')
```

### Univariate bivariate multivariate??
#### Plotting Bivariate Distribution: Joint Plot

#### Bivariate Distribution is used to determine the relation between two variables.

```python
#It gives scatter plot & its strictly bivariate plot
sns.jointplot(x='sepal_length',y='sepal_width',data=i)
```

```python
#use kind= hex
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='hex')
```

```python
#categoriwise scatter plot: make it multivariate
# using an attribute hue
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='scatter',hue='species')
```

```python
#change colors
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='scatter',hue='species',
             palette='afmhot')
```

### Kernel Density Estimation

#### Kernel density estimation is a non-parametric way to estimate the distribution of a variable.

```python
#lets plot kde for all numeric columns
sns.kdeplot(data=i)
```

```python
sns.kdeplot(x='sepal_length',data=i)
```

### Pair relationships using pairplot

#### To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function.

```python
#pairplot works on Numeric data only
sns.pairplot(data=i)
```

```python
#hue is used on categorical data
sns.pairplot(data=i,hue='species')
```

```python
sns.pairplot(data=i[['sepal_length','petal_length','species']],hue='species')
```

```python
sns.pairplot(data=i[['sepal_length','petal_length','species']],hue='species',kind='kde')
```
