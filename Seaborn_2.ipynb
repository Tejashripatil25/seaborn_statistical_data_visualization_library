### Categorical Plotting

#### When one or both the variables under study are categorical, we use plots like stripplot(), swarmplot(), etc,.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Categorical Scatter plots
#### stripplot()
#### stripplot() is used when one of the variable under study is categorical. 
#### It represents the data in sorted order along any one of the axis.

```python
print(sns.get_dataset_names())
```

```python
#load penguin dataset
p = sns.load_dataset('penguins')
p.head()
```

```python
#check unique island
p.island.unique()
```

```python
#check unique species
p.species.unique()
```

```python
#check columns
p.columns
```

### stripplot()

```python
#Draw a scatterplot where one variable is categorical.
# Category shoulb be on x axis
# On the basis of Species differentiate with respect to body mass
#plot body_mass_g w.r.t species
sns.stripplot(x='species',y='body_mass_g',data=p)
```

#### Conclusion: Body mass of Gentoo is higher than other 2 categories, We have More count in Adelie

```python
p.species.value_counts()
```

```python
sns.stripplot(y='species',x='body_mass_g',data=p)
```

```python
#in above plot i want to see genderwise distribustion
sns.stripplot(x='species',y='body_mass_g',data=p,hue='sex',palette='gist_stern')
```

```python
# hue=species
sns.stripplot(x='island',y='body_mass_g',data=p,hue='species',palette='gist_stern')
```

### Box Plots
#### Boxplot is a convenient way to visualize the distribution of data through their quartiles.

```python
# load tips data
t = sns.load_dataset('tips')
t.head()
```

```python
#check unique days
t.day.unique()
```

```python
t.shape
```

```python
#boxplot 
sns.boxplot(x='day',y='total_bill',data=t)
plt.title('Daywise Box plot')
```

```python
#set title('Box plot for Day vs total bill')
ax = sns.boxplot(x='day',y='total_bill',data=t)
#plt.title('Box plot for Day vs total bill')
ax.set_title('Box plot for Day vs total bill')
ax.set_xlabel('Last four days',fontsize=12)
```

```python
#box plot for total_bill
sns.boxplot(data=t.total_bill)
```

```python
sns.boxplot(x='day',y='tip',data=t,hue='smoker',palette='summer')
```

```python
sns.boxplot(x='day',y='tip',data=t,hue='sex',palette='summer')
```

### Bar Plot
#### The barplot() shows the relation between a categorical variable and a continuous variable.
#### The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.

```python
#load titanic data
ti = sns.load_dataset('titanic')
ti.head()
```

```python
ti.shape
```

```python
# check survival rate with respect to gender
sns.barplot(x='sex',y='survived',data=ti)
```

```python
#hue=class
# check survival rate with respect to gender
sns.barplot(x='sex',y='survived',data=ti,hue='class')
```

```python
sns.barplot(x='class',y='survived',data=ti)
```

### count plot
#### A special case in barplot is to show the no of observations in each category rather than
#### computing a statistic for a second variable. For this, we use countplot()

```python
#use titanic dataset
ti.head(2)
```

```python
#check classwise count
sns.countplot(x='class',data=ti)
```

```python
#hue=class
sns.countplot(x='class',data=ti,hue='sex')
```

```python
sns.barplot(x='sex',y='survived',data=ti,hue='class')
```

#### By observing the above result,we can conclude, more male parrengers present than female. More specifically, in third class category we have more male candidates Female are less in count but their survival rate is too high.

### heatmap

#### when we want to show matrix representation with respect to weight of the each feature
#### then go with heatmap

```python
#correlation
cr = ti.corr()
cr
#corr is either +ve or -ve
```

```python
# correlation style
cr.style.background_gradient('twilight')
```

```python
#using heatmap
plt.figure(figsize=(12,10))
sns.heatmap(cr,annot=True)
```

```python
#check corr of few columns
new = ti[['age','fare','pclass']]
cr_2 = new.corr()
cr_2
```

```python
#heatmap
sns.heatmap(cr_2,annot=True,cmap='bone_r')
```
